```python
import numpy as np
import pandas as pd
from ast import literal_eval
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("combined_data.csv")
genreData = df[['title', 'genres']]
genreData.head()
```

```python
genres = pd.read_csv("genre_codes.csv")
genres
```

```python
# Groups from this article: Used clustering
# https://www.scribd.com/document/72221993/Analysis-and-Clustering-of-Movie-Genres
# Used PCA results

GROUP_1 = ["Drama", "Comedy", "Romance", "Family", "Music", "Fantasy", "Animation", "Western"]
GROUP_2 = ["Thriller", "Horror", "Action", "Crime", "Adventure", "Science Fiction", "Mystery"]
GROUP_3 = ["Documentary", "History", "War", "TV Movie"]
```

```python
# Get codes for genres
GROUP_1_codes = list(map(lambda name: genres.loc[genres['name']==name]['id'].values[0], GROUP_1))
GROUP_2_codes = list(map(lambda name: genres.loc[genres['name']==name]['id'].values[0], GROUP_2))
GROUP_3_codes = list(map(lambda name: genres.loc[genres['name']==name]['id'].values[0], GROUP_3))
```

```python
# 50000 in 50 seconds


genreData['group_1'] = 0
genreData['group_2'] = 0
genreData['group_3'] = 0

count = 0
for index, row in genreData.iterrows():
    # print(type(row['genres']))
    genreArray = literal_eval(row['genres'])
    for dict in genreArray:
        # print(type(dict))
        if dict["id"] in GROUP_1_codes:
            genreData.at[count, 'group_1'] = genreData.at[count, 'group_1'] + 1
        elif dict["id"] in GROUP_2_codes:
            genreData.at[count, 'group_2'] = genreData.at[count, 'group_2'] + 1
        elif dict["id"] in GROUP_3_codes:
            genreData.at[count, 'group_3'] = genreData.at[count, 'group_3'] + 1
    if count%5000 == 0: 
        print(count)
    count+=1

print(count)
genreData
```
